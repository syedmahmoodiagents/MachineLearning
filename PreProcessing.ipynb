```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(0)
chi_square_data = np.random.chisquare(df=5, size=100)
```

```python
pd.DataFrame(chi_square_data).plot.kde()
```

```python

sns.kdeplot(chi_square_data)
```

```python
chi_square_data.round(2)
```

```python
from sklearn.preprocessing import PowerTransformer
```

```python
pt = PowerTransformer(method='box-cox')
```

```python
trans = pt.fit_transform(chi_square_data.reshape(-1,1))
```

```python
pd.DataFrame(trans).plot.kde()
```

```python
from pytrends.request import TrendReq

pytrends = TrendReq(
    hl='en-US',   # Language
    tz=330        # Timezone (+5:30 IST)
)
```

```python
pytrends.related_queries()
```

```python

pytrends = TrendReq(hl='en-US', tz=330)
kw_list = ["nike shoes", "bata shoes", "puma shoes"]
pytrends.build_payload(kw_list)

# Get interest over time
data = pytrends.interest_over_time()
# pytrends.related_queries()
print(data)
```
