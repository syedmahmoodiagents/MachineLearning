Recursive Feature Elimination (RFE)

```python
from sklearn.feature_selection import RFE
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
rf = RFE(DecisionTreeClassifier(max_depth=2), n_features_to_select=1)
```

```python
rf.fit(X,Y)
```

```python
rf.ranking_
```

```python
for k in range(4, 0, -1):
    rf = RFE(DecisionTreeClassifier(max_depth=2), n_features_to_select=k)
    rf.fit(X,Y)
    x = rf.transform(X)

    modl = LogisticRegression()
    modl.fit(x, Y)
    print(k, modl.score(x, Y))
```
